```python
import pandas as pd
import jieba,jieba.analyse
from stylecloud import gen_stylecloud
from IPython.display import Image
```

```python
df=pd.read_csv(r'C:\Users\奉华\Desktop\nn\hhk.csv')
df.columns=['标题']
df
```

```python
df['标题']=df['标题'].to_list()
df
```

```python
df2=''.join(df['标题'])
df3=jieba.lcut_for_search(df2)
df4=' '.join(df3)

df5=jieba.analyse.extract_tags(sentence=df4,topK=50,withWeight=False)
df6=' '.join(df5)
df6
cloud=gen_stylecloud(text=df6,font_path=r'C:\Windows\Fonts\simfang.ttf',output_name='text.png',size=400,icon_name='fas fa-ad')
Image('text.png')
```
